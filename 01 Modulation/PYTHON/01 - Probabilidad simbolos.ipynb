```python
import numpy as np

# Probabilidades
p0 = 0.3
p1 = 0.7

# Generar secuencia de 10000 bits
bits = np.random.choice([0, 1], size=10000, p=[p0, p1])


print(bits[:10])

```

```python
grupos = [bits[i:i+6] for i in range(0, len(bits), 2)]
print(grupos[:10])
```

```python
mapping = {
    (0, 0): "a0",
    (0, 1): "a1",
    (1, 1): "a2",
    (1, 0): "a3"
}

simbolos_mapeados = [mapping[tuple(b)] for b in grupos]

print(simbolos_mapeados[:10])
```

```python
from collections import Counter

# Contar ocurrencias
conteo = Counter(simbolos_mapeados)
print(conteo)
# Total de símbolos
total = len(simbolos_mapeados)

# Probabilidades
probabilidades = {k: v/total for k, v in conteo.items()}

print(probabilidades)
```
