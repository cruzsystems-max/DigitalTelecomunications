# Transmisión de Voz con Modulación M-ASK

Sistema completo de transmisión de voz usando modulación M-ASK con canal AWGN

```python
import numpy as np
import matplotlib.pyplot as plt
import sounddevice as sd
from scipy.io import wavfile
from scipy import signal
```

```python
# Parámetros de audio
SAMPLE_RATE = 8000
CHANNELS = 1
BITS_PER_SAMPLE = 8

# Parámetros de modulación M-ASK
M = 16
n = int(np.log2(M))
formato = 'polar'
Ac = 1

# Probabilidad de bits
p0 = 0.5
p1 = 1 - p0

# SNR a probar
SNR_VALUES = [5, 10, 15, 20, 25]

print(f"Parámetros de Audio:")
print(f"  Frecuencia de muestreo: {SAMPLE_RATE} Hz")
print(f"  Cuantización: {BITS_PER_SAMPLE} bits")
print(f"\nParámetros de Modulación:")
print(f"  M = {M}")
print(f"  Formato: {formato}")
print(f"  SNR a probar: {SNR_VALUES} dB")
```

```python
# Funciones auxiliares para código Gray

def gray_code(n):
    if n == 0:
        return ['']
    prev = gray_code(n-1)
    return ['0' + code for code in prev] + ['1' + code for code in reversed(prev)]

gray = gray_code(n)
print(f"Códigos Gray para M={M}:")
for i, codigo in enumerate(gray):
    print(f"  {i:2d}: {codigo}")
```

```python
# Funciones de captura de audio

def record_audio():
    print("Grabando audio... Presiona Enter para detener.")
    recording = []
    
    def callback(indata, frames, time, status):
        if status:
            print(status)
        recording.append(indata.copy())
    
    stream = sd.InputStream(samplerate=SAMPLE_RATE, channels=CHANNELS, callback=callback, dtype='float32')
    with stream:
        input()
    
    audio = np.concatenate(recording, axis=0).flatten()
    print(f"Grabación completa: {len(audio)} muestras ({len(audio)/SAMPLE_RATE:.2f} segundos)")
    return audio

def load_audio(filepath):
    sample_rate, audio = wavfile.read(filepath)
    
    if audio.dtype == np.int16:
        audio = audio.astype(float) / 32768.0
    elif audio.dtype == np.uint8:
        audio = (audio.astype(float) / 255.0) * 2.0 - 1.0
    
    if len(audio.shape) > 1:
        audio = audio.mean(axis=1)
    
    if sample_rate != SAMPLE_RATE:
        from scipy.signal import resample
        num_samples = int(len(audio) * SAMPLE_RATE / sample_rate)
        audio = resample(audio, num_samples)
        print(f"Audio remuestreado de {sample_rate} Hz a {SAMPLE_RATE} Hz")
    
    print(f"Audio cargado: {len(audio)} muestras ({len(audio)/SAMPLE_RATE:.2f} segundos)")
    return audio

print("Funciones de captura definidas.")
```

```python
# Captura de audio

choice = input("¿Grabar audio (G) o Cargar archivo WAV (C)? ").strip().upper()

if choice == 'G':
    audio_original = record_audio()
elif choice == 'C':
    filepath = input("Ingresa la ruta del archivo WAV: ").strip()
    audio_original = load_audio(filepath)
else:
    print("Opción inválida. Usando audio de ejemplo (tono 440 Hz, 3 seg)")
    duration = 3.0
    t = np.arange(int(duration * SAMPLE_RATE)) / SAMPLE_RATE
    audio_original = 0.5 * np.sin(2 * np.pi * 440 * t)

print(f"\nAudio capturado: {len(audio_original)} muestras ({len(audio_original)/SAMPLE_RATE:.2f} seg)")
```

```python
# Visualización del audio original

time = np.arange(len(audio_original)) / SAMPLE_RATE

fig, axes = plt.subplots(2, 1, figsize=(12, 6))

axes[0].plot(time, audio_original, linewidth=0.5)
axes[0].set_title('Audio Original - Forma de Onda')
axes[0].set_xlabel('Tiempo (s)')
axes[0].set_ylabel('Amplitud')
axes[0].grid(True, alpha=0.3)

f, t, Sxx = signal.spectrogram(audio_original, SAMPLE_RATE, nperseg=256)
axes[1].pcolormesh(t, f, 10*np.log10(Sxx + 1e-10), shading='gouraud', cmap='viridis')
axes[1].set_title('Audio Original - Espectrograma')
axes[1].set_ylabel('Frecuencia (Hz)')
axes[1].set_xlabel('Tiempo (s)')

plt.tight_layout()
plt.show()
```

```python
# Conversión de audio a bits

audio_clipped = np.clip(audio_original, -1.0, 1.0)
audio_uint8 = ((audio_clipped + 1.0) / 2.0 * 255).astype(np.uint8)
bits_tx = np.unpackbits(audio_uint8)

print(f"Audio cuantizado: {len(audio_uint8)} muestras de 8 bits")
print(f"Stream de bits: {len(bits_tx)} bits")
print(f"Símbolos M-ASK: {len(bits_tx)//n} símbolos")
print(f"\nPrimeras 10 muestras de audio (uint8): {audio_uint8[:10]}")
print(f"Primeros 40 bits: {bits_tx[:40]}")
```

```python
# Modulación M-ASK (Convertidor de datos)

def convertidor_datos(bits, M=4, formato='polar'):
    n = int(np.log2(M))
    
    L = len(bits)
    if L % n != 0:
        bits = bits[:L - (L % n)]
    
    grupos = bits.reshape(-1, n)
    gray = gray_code(n)
    gray_array = np.array([[int(b) for b in codigo] for codigo in gray])
    indices = np.where((grupos[:, None, :] == gray_array[None, :, :]).all(axis=2))[1]
    
    if formato == 'polar':
        niveles = 2*indices - (M - 1)
    elif formato == 'unipolar':
        niveles = indices
    else:
        raise ValueError("Formato debe ser 'polar' o 'unipolar'")
    
    return niveles

simbolos_tx = convertidor_datos(bits_tx, M, formato)

print(f"Símbolos transmitidos: {len(simbolos_tx)}")
print(f"Primeros 20 símbolos: {simbolos_tx[:20]}")
```

```python
# Cálculo de parámetros para canal AWGN

p_simbolos = np.ones(M) / M

indices_niveles = np.arange(M)
if formato == 'polar':
    niveles_ask = 2*indices_niveles - (M - 1)
elif formato == 'unipolar':
    niveles_ask = indices_niveles

ak2_mean = np.sum(niveles_ask**2 * p_simbolos)

print(f"Niveles M-ASK: {niveles_ask}")
print(f"Valor cuadrático medio (a_k^2)̅ = {ak2_mean:.6f}")
```

```python
# Canal AWGN (múltiples SNRs)

resultados_canal = {}

for snr_db in SNR_VALUES:
    gamma_b_lineal = 10**(snr_db / 10)
    sigma2 = (Ac**2 * ak2_mean) / (2 * gamma_b_lineal * np.log2(M))
    sigma = np.sqrt(sigma2)
    
    ni = np.random.normal(0, sigma, size=len(simbolos_tx))
    yi = simbolos_tx + ni
    
    resultados_canal[snr_db] = {
        'yi': yi,
        'ni': ni,
        'sigma': sigma,
        'sigma2': sigma2
    }
    
    print(f"SNR = {snr_db:2d} dB: σ² = {sigma2:.6f}, σ = {sigma:.6f}")

print(f"\nSeñal recibida generada para {len(SNR_VALUES)} valores de SNR")
```

```python
# Cuantización (Demodulación)

umbrales = (niveles_ask[:-1] + niveles_ask[1:]) / 2

print(f"Umbrales de cuantización: {umbrales}")

for snr_db in SNR_VALUES:
    yi = resultados_canal[snr_db]['yi']
    indices_rx = np.digitize(yi, umbrales)
    simbolos_rx = niveles_ask[indices_rx]
    
    resultados_canal[snr_db]['simbolos_rx'] = simbolos_rx
    
    errores_simbolo = np.sum(simbolos_tx != simbolos_rx)
    print(f"SNR = {snr_db:2d} dB: {errores_simbolo} errores de símbolo de {len(simbolos_tx)}")
```

```python
# Demapeo de símbolos a bits

def demapeador_simbolos(simbolos, M=4, formato='polar'):
    n = int(np.log2(M))
    gray = gray_code(n)
    
    if formato == 'polar':
        indices = ((simbolos + (M - 1)) / 2).astype(int)
    elif formato == 'unipolar':
        indices = simbolos.astype(int)
    else:
        raise ValueError("Formato debe ser 'polar' o 'unipolar'")
    
    lookup_table = np.array([[int(b) for b in codigo] for codigo in gray], dtype=np.int8)
    
    chunk_size = 1000000
    num_chunks = (len(indices) + chunk_size - 1) // chunk_size
    
    bits_estimados = np.empty(len(indices) * n, dtype=np.int8)
    
    for i in range(num_chunks):
        start = i * chunk_size
        end = min((i + 1) * chunk_size, len(indices))
        chunk_indices = indices[start:end]
        
        bits_chunk = lookup_table[chunk_indices].flatten()
        bits_estimados[start*n:end*n] = bits_chunk
    
    return bits_estimados

for snr_db in SNR_VALUES:
    simbolos_rx = resultados_canal[snr_db]['simbolos_rx']
    bits_rx = demapeador_simbolos(simbolos_rx, M, formato)
    
    resultados_canal[snr_db]['bits_rx'] = bits_rx
    
    bits_tx_truncated = bits_tx[:len(bits_rx)]
    BER = np.sum(bits_tx_truncated != bits_rx) / len(bits_rx)
    resultados_canal[snr_db]['BER'] = BER
    
    print(f"SNR = {snr_db:2d} dB: BER = {BER:.4e}")
```

```python
# Reconstrucción de audio

resultados_audio = {}

for snr_db in SNR_VALUES:
    bits_rx = resultados_canal[snr_db]['bits_rx']
    
    bits_padded = bits_rx[:len(audio_original) * 8]
    audio_rx_uint8 = np.packbits(bits_padded)
    audio_rx_float = (audio_rx_uint8.astype(float) / 255.0) * 2.0 - 1.0
    audio_rx_float = np.clip(audio_rx_float, -1.0, 1.0)
    
    resultados_audio[snr_db] = {
        'audio': audio_rx_float,
        'BER': resultados_canal[snr_db]['BER']
    }
    
    print(f"SNR = {snr_db:2d} dB: Audio reconstruido ({len(audio_rx_float)} muestras)")

print(f"\nReconstrucción completa para {len(SNR_VALUES)} valores de SNR")
```

```python
# Visualización comparativa

num_snr = len(SNR_VALUES)
fig, axes = plt.subplots(num_snr + 1, 2, figsize=(14, 3*(num_snr+1)))

time = np.arange(len(audio_original)) / SAMPLE_RATE

axes[0, 0].plot(time, audio_original, linewidth=0.5)
axes[0, 0].set_title('Audio Original - Forma de Onda')
axes[0, 0].set_xlabel('Tiempo (s)')
axes[0, 0].set_ylabel('Amplitud')
axes[0, 0].grid(True, alpha=0.3)

f, t, Sxx = signal.spectrogram(audio_original, SAMPLE_RATE, nperseg=256)
axes[0, 1].pcolormesh(t, f, 10*np.log10(Sxx + 1e-10), shading='gouraud', cmap='viridis')
axes[0, 1].set_title('Audio Original - Espectrograma')
axes[0, 1].set_ylabel('Frecuencia (Hz)')
axes[0, 1].set_xlabel('Tiempo (s)')

for idx, snr_db in enumerate(sorted(SNR_VALUES), start=1):
    audio_rx = resultados_audio[snr_db]['audio']
    ber = resultados_audio[snr_db]['BER']
    
    axes[idx, 0].plot(time, audio_rx, linewidth=0.5, color='orange')
    axes[idx, 0].set_title(f'SNR = {snr_db} dB - BER = {ber:.4e}')
    axes[idx, 0].set_xlabel('Tiempo (s)')
    axes[idx, 0].set_ylabel('Amplitud')
    axes[idx, 0].grid(True, alpha=0.3)
    
    f, t, Sxx = signal.spectrogram(audio_rx, SAMPLE_RATE, nperseg=256)
    axes[idx, 1].pcolormesh(t, f, 10*np.log10(Sxx + 1e-10), shading='gouraud', cmap='viridis')
    axes[idx, 1].set_ylabel('Frecuencia (Hz)')
    axes[idx, 1].set_xlabel('Tiempo (s)')

plt.tight_layout()
plt.show()
```

```python
# Gráfica BER vs SNR

from scipy.special import erfc

snr_values = sorted(SNR_VALUES)
ber_simulado = [resultados_audio[snr]['BER'] for snr in snr_values]

L = np.log2(M)
ber_teorico = []
for snr_db in snr_values:
    gamma_b_lin = 10**(snr_db / 10)
    argumento = np.sqrt((6 * L / (M**2 - 1)) * gamma_b_lin)
    Q = 0.5 * erfc(argumento / np.sqrt(2))
    pb = (2 / L) * (1 - 1/M) * Q
    ber_teorico.append(pb)

plt.figure(figsize=(10, 6))
plt.semilogy(snr_values, ber_simulado, 'bo-', linewidth=2, markersize=8, label='BER Simulado - Voz')
plt.semilogy(snr_values, ber_teorico, 'r--', linewidth=2, label='BER Teórico')
plt.xlabel('SNR por bit (dB)')
plt.ylabel('BER')
plt.title(f'Transmisión de Voz - BER vs SNR (M={M}, {formato})')
plt.grid(True, which='both', alpha=0.3)
plt.legend()
plt.xlim([snr_values[0], snr_values[-1]])
plt.show()

print(f"\n{'SNR (dB)':<10} {'BER Simulado':<15} {'BER Teórico':<15}")
print("-" * 45)
for snr, ber_s, ber_t in zip(snr_values, ber_simulado, ber_teorico):
    print(f"{snr:<10d} {ber_s:<15.6e} {ber_t:<15.6e}")
```

```python
# Reproducción de audio

import os

output_dir = './voice_results'
os.makedirs(output_dir, exist_ok=True)

wavfile.write(f"{output_dir}/original.wav", SAMPLE_RATE, (audio_original * 32767).astype(np.int16))
print(f"Audio original guardado: {output_dir}/original.wav")

print("\n" + "="*60)
print("REPRODUCCIÓN DE AUDIO")
print("="*60)

print("\nReproduciendo audio ORIGINAL...")
sd.play(audio_original, SAMPLE_RATE)
sd.wait()
input("Presiona Enter para continuar...")

for snr_db in sorted(SNR_VALUES):
    audio_rx = resultados_audio[snr_db]['audio']
    ber = resultados_audio[snr_db]['BER']
    
    print(f"\nReproduciendo SNR = {snr_db} dB (BER = {ber:.4e})")
    sd.play(audio_rx, SAMPLE_RATE)
    sd.wait()
    
    filename = f"{output_dir}/snr_{snr_db}dB_BER_{ber:.2e}.wav"
    wavfile.write(filename, SAMPLE_RATE, (audio_rx * 32767).astype(np.int16))
    print(f"Guardado: {filename}")
    
    input("Presiona Enter para siguiente...")

print("\n" + "="*60)
print("TRANSMISIÓN COMPLETADA")
print("="*60)
```
